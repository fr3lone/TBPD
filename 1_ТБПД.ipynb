```python
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', 100)
plt.style.use('seaborn-v0_8-darkgrid')
sns.set_palette("husl")

print("Библиотеки успешно импортированы!")
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv')

print("Данные успешно загружены!")
print(f"Размерность данных: {df.shape}")
print(f"Количество строк: {df.shape[0]}")
print(f"Количество столбцов: {df.shape[1]}")

print("ПЕРВЫЕ 5 СТРОК ДАННЫХ:")
display(df.head())

print("\nПОСЛЕДНИЕ 5 СТРОК ДАННЫХ:")
display(df.tail())

print("\nИНФОРМАЦИЯ О ДАННЫХ:")
print(df.info())

print("\nСТАТИСТИЧЕСКОЕ ОПИСАНИЕ:")
display(df.describe())

print("\nТИПЫ ДАННЫХ:")
print(df.dtypes)
```

```python
# 4.1 Проверка пропущенных значений
print("ПРОПУЩЕННЫЕ ЗНАЧЕНИЯ:")
print(df.isnull().sum())
print(f"Процент пропусков: {(df.isnull().sum() / len(df) * 100).round(2)}%")

# 4.2 Проверка дубликатов
print(f"\nКОЛИЧЕСТВО ДУБЛИКАТОВ: {df.duplicated().sum()}")

# 4.3 Проверка уникальных значений в категориальных столбцах
categorical_cols = df.select_dtypes(include=['object']).columns
for col in categorical_cols:
    print(f"\nУНИКАЛЬНЫЕ ЗНАЧЕНИЯ В СТОЛБЦЕ '{col}':")
    print(df[col].value_counts())

# 4.4 Обработка выбросов (метод IQR)
numeric_cols = df.select_dtypes(include=[np.number]).columns

for col in numeric_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
    print(f"\nСтолбец '{col}':")
    print(f"  Нижняя граница: {lower_bound:.2f}")
    print(f"  Верхняя граница: {upper_bound:.2f}")
    print(f"  Количество выбросов: {len(outliers)}")
```

```python
# 5.1 Общая статистика по числовым переменным
print("ОСНОВНЫЕ СТАТИСТИКИ:")
for col in numeric_cols:
    print(f"\n{col}:")
    print(f"  Среднее: {df[col].mean():.2f}")
    print(f"  Медиана: {df[col].median():.2f}")
    print(f"  Мода: {df[col].mode()[0]}")
    print(f"  Стандартное отклонение: {df[col].std():.2f}")
    print(f"  Дисперсия: {df[col].var():.2f}")

# 5.2 Анализ по категориям
for cat_col in categorical_cols:
    print(f"\nАНАЛИЗ ПО {cat_col.upper()}:")
    print(df.groupby(cat_col)[numeric_cols].mean().round(2))

# 5.3 Корреляционный анализ
print("\nКОРРЕЛЯЦИОННАЯ МАТРИЦА:")
correlation_matrix = df[numeric_cols].corr()
display(correlation_matrix.round(2))
```

```python
# Создаем фигуру с несколькими подграфиками
fig = plt.figure(figsize=(16, 12))

# 6.1 Гистограммы распределения числовых переменных
for i, col in enumerate(numeric_cols, 1):
    plt.subplot(3, 3, i)
    df[col].hist(bins=20, edgecolor='black', alpha=0.7)
    plt.title(f'Распределение {col}')
    plt.xlabel(col)
    plt.ylabel('Частота')
plt.tight_layout()
plt.show()
```

```python
# 6.2 Ящики с усами (box plots) для выявления выбросов
fig, axes = plt.subplots(1, len(numeric_cols), figsize=(15, 5))
for i, col in enumerate(numeric_cols):
    df.boxplot(column=col, ax=axes[i])
    axes[i].set_title(f'Box plot: {col}')
plt.tight_layout()
plt.show()
```

```python
# 6.3 Корреляционная матрица (тепловая карта)
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', 
            center=0, square=True, linewidths=1)
plt.title('Тепловая карта корреляций')
plt.tight_layout()
plt.show()
```

```python
# 6.4 Столбчатые диаграммы для категориальных переменных
fig, axes = plt.subplots(1, len(categorical_cols), figsize=(15, 5))
if len(categorical_cols) == 1:
    df[categorical_cols[0]].value_counts().plot(kind='bar', ax=axes)
    axes.set_title(f'Распределение {categorical_cols[0]}')
    axes.set_xlabel(categorical_cols[0])
    axes.set_ylabel('Количество')
else:
    for i, col in enumerate(categorical_cols):
        df[col].value_counts().plot(kind='bar', ax=axes[i])
        axes[i].set_title(f'Распределение {col}')
        axes[i].set_xlabel(col)
        axes[i].set_ylabel('Количество')
plt.tight_layout()
plt.show()
```

```python
# Анализ зависимостей между категориальными и числовыми переменными
for cat_col in categorical_cols:
    fig, axes = plt.subplots(1, len(numeric_cols), figsize=(15, 4))
    for i, num_col in enumerate(numeric_cols):
        df.boxplot(column=num_col, by=cat_col, ax=axes[i])
        axes[i].set_title(f'{num_col} по {cat_col}')
        axes[i].set_xlabel(cat_col)
    plt.suptitle(f'Анализ зависимости от {cat_col}', y=1.02)
    plt.tight_layout()
    plt.show()
```

```python
print("ВЫВОДЫ ПО РЕЗУЛЬТАТАМ АНАЛИЗА:")
print("="*50)

# Основные наблюдения
print("\n1. ОСНОВНЫЕ ХАРАКТЕРИСТИКИ ДАННЫХ:")
print(f"   - Объем данных: {df.shape[0]} записей, {df.shape[1]} признаков")
print(f"   - Типы данных: {df.dtypes.value_counts().to_dict()}")

# Качество данных
print("\n2. КАЧЕСТВО ДАННЫХ:")
print(f"   - Пропуски: {'отсутствуют' if df.isnull().sum().sum() == 0 else 'обнаружены'}")
print(f"   - Дубликаты: {'отсутствуют' if df.duplicated().sum() == 0 else 'обнаружены'}")

# Ключевые статистики
print("\n3. КЛЮЧЕВЫЕ СТАТИСТИКИ:")
for col in numeric_cols:
    print(f"   - {col}: среднее={df[col].mean():.2f}, медиана={df[col].median():.2f}")

# Корреляции
print("\n4. СИЛЬНЫЕ КОРРЕЛЯЦИИ:")
strong_corr = []
for i in range(len(correlation_matrix.columns)):
    for j in range(i+1, len(correlation_matrix.columns)):
        if abs(correlation_matrix.iloc[i, j]) > 0.5:
            strong_corr.append(f"{correlation_matrix.columns[i]} - {correlation_matrix.columns[j]}: {correlation_matrix.iloc[i, j]:.2f}")
if strong_corr:
    for corr in strong_corr:
        print(f"   - {corr}")
else:
    print("   - Сильных корреляций не обнаружено")
```
